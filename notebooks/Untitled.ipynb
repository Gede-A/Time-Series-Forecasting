```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the tickers and date range
tickers = ['TSLA', 'BND', 'SPY']
start_date = '2015-01-01'
end_date = '2024-10-31'

# Download the data
data = yf.download(tickers, start=start_date, end=end_date, interval='1d')

data.to_csv('./data/financial_data.csv')

# Display the first few rows
data.head()
```

```python
framed_data = pd.DataFrame(data)

# Iterate through columns and print data types
for column in framed_data.columns:
    print(f"{column}: {framed_data[column].dtype}")
```

```python
print(framed_data.isnull().sum())
```

```python
for column in framed_data.columns:
    print(f'mean of {framed_data[column]}: {framed_data[column].mean()}')
```
