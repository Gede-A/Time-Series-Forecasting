```python
# Calculate daily percentage change
daily_returns = framed_data_normalized['Close'].pct_change() * 100  # Daily percentage change in Close prices

# Define the rolling window
rolling_window = 21

# Calculate rolling mean and rolling standard deviation
rolling_mean = daily_returns.rolling(window=rolling_window).mean()
rolling_std = daily_returns.rolling(window=rolling_window).std()

# Plot the rolling mean and standard deviation
plt.figure(figsize=(14, 10))
for ticker in ['BND', 'SPY', 'TSLA']:
    plt.plot(daily_returns.index, daily_returns[ticker], label=f"{ticker} Daily Returns")
    plt.plot(rolling_mean.index, rolling_mean[ticker], label=f"{ticker} {rolling_window}-Day Rolling Mean")
    plt.plot(rolling_std.index, rolling_std[ticker], label=f"{ticker} {rolling_window}-Day Rolling Std Dev")
    plt.title(f'{ticker} Rolling Mean and Volatility')
    plt.xlabel('Date')
    plt.ylabel('Percentage Change (%)')
    plt.legend(loc='upper left')
    plt.show()

# Detect outliers: returns beyond 2 standard deviations from the rolling mean
outliers = (daily_returns - rolling_mean).abs() > (2 * rolling_std)

# Display days with unusually high or low returns (outliers)
for ticker in ['BND', 'SPY', 'TSLA']:
    ticker_outliers = daily_returns[ticker][outliers[ticker]]
    print(f"\nOutliers for {ticker}:\n", ticker_outliers.dropna())

# Analyze days with unusually high and low returns
high_returns = daily_returns.idxmax()
low_returns = daily_returns.idxmin()

print("\nDays with highest returns for each ticker:")
for ticker in ['BND', 'SPY', 'TSLA']:
    print(f"{ticker}: {high_returns[ticker]} - {daily_returns[ticker].loc[high_returns[ticker]]}%")

print("\nDays with lowest returns for each ticker:")
for ticker in ['BND', 'SPY', 'TSLA']:
    print(f"{ticker}: {low_returns[ticker]} - {daily_returns[ticker].loc[low_returns[ticker]]}%")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Resample to monthly data for clearer seasonality analysis (optional)
monthly_data = data['Adj Close'].resample('M').mean()

# Perform seasonal decomposition for each ticker using the additive model
for ticker in ['BND', 'SPY', 'TSLA']:
    print(f"\nDecomposing {ticker} Time Series\n")
    
    # Decompose the time series into trend, seasonal, and residuals (using additive model)
    decomposition = seasonal_decompose(monthly_data[ticker], model='additive', period=12)
    
    # Plot the decomposition
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
    decomposition.observed.plot(ax=ax1, title=f"{ticker} Observed", legend=False)
    decomposition.trend.plot(ax=ax2, title="Trend", legend=False)
    decomposition.seasonal.plot(ax=ax3, title="Seasonal", legend=False)
    decomposition.resid.plot(ax=ax4, title="Residual", legend=False)
    plt.tight_layout()
    plt.show()
```
